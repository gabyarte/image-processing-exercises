```python
import cv2

import matplotlib.pyplot as plt
```

```python
def show_image(image, name='Lenna'):
    plt.imshow(image)
    plt.title(name)
    plt.show()
```

```python
image = cv2.imread('../images/Lenna.png')
grayed_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

show_image(image)
print(image.shape)
```

## SIFT

```python
sift = cv2.SIFT_create()
sift_key_points = sift.detect(grayed_image, None)

sift_keypointed_image = cv2.drawKeypoints(grayed_image, sift_key_points, image)
cv2.imwrite('sift_keypoints.jpg', sift_keypointed_image)
show_image(sift_keypointed_image, 'Image with SIFT keypoints')
```

## BRIEF

```python
star = cv2.xfeatures2d.StarDetector_create()
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()
brief_key_points = star.detect(grayed_image, None)
brief_key_points, des = brief.compute(grayed_image, brief_key_points)

brief_keypointed_image = cv2.drawKeypoints(grayed_image, brief_key_points, image)
cv2.imwrite('brief_keypoints.jpg', brief_keypointed_image)
show_image(brief_keypointed_image, 'Image with BRIEF keypoints')
```

## BRISK

```python
brisk = cv2.BRISK_create()

brisk_key_points, _ = brisk.detectAndCompute(grayed_image, None)
brisk_keypointed_image = cv2.drawKeypoints(grayed_image, brisk_key_points, image)
cv2.imwrite('brisk_keypoints.jpg', brisk_keypointed_image)
show_image(brisk_keypointed_image, 'Image with BRISK keypoints')
```
